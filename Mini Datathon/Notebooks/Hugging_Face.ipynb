# 03 - Model

## 0301 - Download specific model

```python
pipeline_model = pipeline(model='siebert/sentiment-roberta-large-english')
pipeline_model
```

## 0302 - Use model

```python
pipeline_model("hi there, beutiful")
```

```python
pipeline_model("shut up and leave")
```

## 0303 - Use model with several messages

```python
df = pd.DataFrame({"chat-messages":["ths is the best!", "leave me alone"]})
df
```

```python
pipeline_model(list(df["chat-messages"]))
```

```python
pd.DataFrame(pipeline_model(list(df["chat-messages"])))
```

```python
df_pred = pd.concat([df, pd.DataFrame(pipeline_model(list(df["chat-messages"])))], axis=1)
df_pred
```
